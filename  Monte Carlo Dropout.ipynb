Enable dropout at inference

```python
class ProbabilisticHead(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.fc = nn.Linear(dim, 1)
        self.dropout = nn.Dropout(p=0.3)

    def forward(self, x):
        x = self.dropout(x)
        return self.fc(x)
```

Inference

```python
def mc_predict(model, x, runs=30):
    model.train()  # IMPORTANT
    preds = torch.stack([model(x) for _ in range(runs)])
    mean = preds.mean(0)
    std  = preds.std(0)
    return mean, std
```



    Mean → expected traffic

    Std → uncertainty (risk)
